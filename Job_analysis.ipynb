```python
def remove_outliers_iqr(df, columns):
    Q1 = df[columns].quantile(0.25)
    Q3 = df[columns].quantile(0.75)
    IQR = Q3 - Q1

    filter_ = ~((df[columns] < (Q1 - 1.5 * IQR)) | (df[columns] > (Q3 + 1.5 * IQR))).any(axis=1)

    return df[filter_]

independent_vars = df.drop(columns=['PROFILE', 'Unnamed: 0', 'NAME', 'USER_ID']).columns

df_no_outliers = remove_outliers_iqr(df, independent_vars)

print("Number of rows before removing outliers:", df.shape[0])
print("Number of rows after removing outliers:", df_no_outliers.shape[0])

num_plots = len(independent_vars)
rows = (num_plots + 2) // 3
cols = 3

fig, axes = plt.subplots(rows, cols, figsize=(15, 5*rows))

axes = axes.flatten()

for i, column in enumerate(independent_vars):
    sns.boxplot(x=df_no_outliers[column], ax=axes[i])
    axes[i].set_title(column)

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

def plot_decision_boundary_svm(X, y, model):
    # Define the colormap using ListedColormap
    cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
    cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

    # Define the meshgrid for plotting decision boundary
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))

    # Predict using the model for each point in the meshgrid
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot the decision boundary using contourf with the light colormap
    plt.contourf(xx, yy, Z, cmap=cmap_light, alpha=0.3)

    # Scatter plot for the actual data points with the bold colormap
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold, edgecolor='k')

    # Add titles and labels
    plt.title("Decision Boundary SVM")
    plt.xlabel("PCA Component 1")
    plt.ylabel("PCA Component 2")

    # Display the plot
    plt.show()

# Example usage:
plot_decision_boundary_svm(x_train_pca, y_train_encoded, svm_model)
```

```python
from bayes_opt import BayesianOptimization

def rf_cv(n_estimators, max_depth, min_samples_split):
    n_estimators = int(n_estimators)
    max_depth = int(max_depth)
    min_samples_split = int(min_samples_split)

    model = RandomForestClassifier(
        n_estimators=n_estimators,
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        random_state=42
    )

    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)

    accuracy = accuracy_score(y_test, y_pred)

    return accuracy
```

```python
pbounds = {
    'n_estimators': (10, 200),
    'max_depth': (1, 20),
    'min_samples_split': (2, 10)
}

optimizer = BayesianOptimization(
    f=rf_cv,
    pbounds=pbounds,
    random_state=42,
    verbose=2
)
optimizer.maximize(
    init_points=10,
    n_iter=30
)
```

```python
best_params_rf = optimizer.max['params']
best_params_rf['n_estimators'] = int(best_params_rf['n_estimators'])
best_params_rf['max_depth'] = int(best_params_rf['max_depth'])
best_params_rf['min_samples_split'] = int(best_params_rf['min_samples_split'])

print(f"Best hyperparameters found: {best_params_rf}")


best_model_rf = RandomForestClassifier(
    n_estimators=best_params_rf['n_estimators'],
    max_depth=best_params_rf['max_depth'],
    min_samples_split=best_params_rf['min_samples_split'],
    random_state=42
)

best_model_rf.fit(x_train, y_train)
y_pred_best_rf = best_model_rf.predict(x_test)

accuracy_best_rf = accuracy_score(y_test, y_pred_best_rf)
print(f"Accuracy of Best Random Forest using Bayesian Optimization: {accuracy_best_rf:.3f}")

report = classification_report(y_test, y_pred_best_rf)
print("Classification Report:\n")
print(report)
```

```python
import random
from sklearn.model_selection import cross_val_score
from deap import base, creator, tools, algorithms

def random_forest_hyperparameters():
    n_estimators = random.randint(10, 200)
    max_depth = random.randint(1, 30)
    min_samples_split = random.randint(2, 10)
    return n_estimators, max_depth, min_samples_split

def evaluate(individual):
    n_estimators, max_depth, min_samples_split = individual
    model = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth, min_samples_split=min_samples_split, random_state=42)
    scores = cross_val_score(model, x_train, y_train, cv=4, scoring='accuracy')
    return scores.mean(),

creator.create("FitnessMax", base.Fitness, weights=(1.0,))
creator.create("Individual", list, fitness=creator.FitnessMax)

toolbox = base.Toolbox()
toolbox.register("attr_int", random_forest_hyperparameters)
toolbox.register("individual", tools.initIterate, creator.Individual, toolbox.attr_int)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

toolbox.register("mate", tools.cxUniform, indpb=0.5)
toolbox.register("mutate", tools.mutUniformInt, low=[10, 1, 2], up=[200, 30, 10], indpb=0.2)
toolbox.register("select", tools.selTournament, tournsize=3)
toolbox.register("evaluate", evaluate)

population = toolbox.population(n=20)
ngen = 10
cxpb = 0.5
mutpb = 0.2

best = algorithms.eaSimple(population, toolbox, cxpb, mutpb, ngen, verbose=True)

best_individual = tools.selBest(population, k=1)[0]
print(f'Best Individual: {best_individual}')
print(f'Best Fitness: {best_individual.fitness.values[0]}')
```
